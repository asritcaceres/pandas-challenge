```python
# Dependencies and Setup
import pandas as pd
# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
#Total number of players
unique_players = len(purchase_data["SN"].unique())

Total_players = pd.DataFrame({"Total Players": [unique_players]})
Total_players
```

```python
unique_items = len(purchase_data["Item ID"].unique ())
total_revenue = purchase_data["Price"].sum()
total_num_purchases = purchase_data["Purchase ID"].count()
average_price = total_revenue/total_num_purchases
information_df = pd.DataFrame({"Total Unique Item": [unique_items], "Average Price": [average_price], "Total Number of Purchases": [total_num_purchases], "Total Revenue": [total_revenue]})
information_df["Average Price"]= information_df["Average Price"].map("${:.2f}".format)
information_df["Total Revenue"]= information_df["Total Revenue"].map("${:.2f}".format)
information_df
```

```python
male_players = purchase_data.loc[purchase_data["Gender"]=="Male"]
male_count = len(male_players["SN"].unique())
percent_male = "{:.2f}%".format(male_count/unique_players *100)

female_players = purchase_data.loc[purchase_data["Gender"]=="Female"]
female_count = len(female_players["SN"].unique())
percent_female = "{:.2f}%".format(female_count/unique_players *100)


other_players = purchase_data.loc[purchase_data["Gender"]=="Other / Non-Disclosed"]
other_count = len(other_players["SN"].unique())
percent_other = "{:.2f}%".format(other_count/unique_players *100)

gender_demographics = pd.DataFrame([{"Gender": "Male", "Total Count": male_count, "Total Percent": percent_male},{"Gender": "Female", "Total Count": female_count, "Total Percent": percent_female},{"Gender": "Other/Non-Disclosed", "Total Count": other_count, "Total Percent": percent_other}])
gender_demographics
```

```python
#purchase count by gender
male_purchase_data = purchase_data.loc[purchase_data["Gender"]== "Male", :]
male_purchase_count = len(male_purchase_data)

female_purchase_data = purchase_data.loc[purchase_data["Gender"]== "Female", :]
female_purchase_count = len(female_purchase_data)
                          
other_purchase_data = purchase_data.loc[purchase_data["Gender"]== "Other / Non-Disclosed", :]
other_purchase_count = len(other_purchase_data)


#Avg purchase price by gender
male_avg_purchase = purchase_data.loc[purchase_data["Gender"]== "Male", ["Price"]].mean()
female_avg_purchase = purchase_data.loc[purchase_data["Gender"]== "Female", ["Price"]].mean()
other_avg_purchase = purchase_data.loc[purchase_data["Gender"]== "Other / Non-Disclosed", ["Price"]].mean()

#total purchase price by gender
male_total_purchase_price = purchase_data.loc[purchase_data["Gender"]== "Male", ["Price"]].sum()
female_total_purchase_price = purchase_data.loc[purchase_data["Gender"]== "Female", ["Price"]].sum()
other_total_purchase_price = purchase_data.loc[purchase_data["Gender"]== "Other / Non-Disclosed", ["Price"]].sum()

#avg purchase total by gender
male_avg_total_purchase = male_total_purchase_price/male_count
female_avg_total_purchase = female_total_purchase_price/female_count
other_avg_total_purchase = other_total_purchase_price/other_count

#Dataframe 
purchase_info_gender = pd.DataFrame([{"Gender": "Male", "Total Purchase Count": male_purchase_count, "Average Purchase Price": "${:.2f}".format(male_avg_purchase[0]), "Total Purchase Price":"${:.2f}".format(male_total_purchase_price[0]), "Average Total Purchase":"${:.2f}".format(male_avg_total_purchase[0])},{"Gender": "Female", "Total Purchase Count": female_purchase_count, "Average Purchase Price": "${:.2f}".format(female_avg_purchase[0]), "Total Purchase Price": "${:.2f}".format(female_total_purchase_price[0]), "Average Total Purchase": "${:.2f}".format(female_avg_total_purchase[0])},{"Gender": "Other/Non-Disclosed", "Total Purchase Count": other_purchase_count, "Average Purchase Price":"${:.2f}".format(other_avg_purchase[0]), "Total Purchase Price": "${:.2f}".format(other_total_purchase_price[0]), "Average Total Purchase": "${:.2f}".format(other_avg_total_purchase[0])}])
purchase_info_gender


```

```python
#Establish bins for ages
bins = [0, 9, 14, 19, 24, 29, 34, 39, 46]
age_names = ["0-9","10-14","15-19","20-24","25-29","30-34","35-39", "40-46"]
pd.cut(purchase_data["Age"], bins, labels=age_names).head()

#Create a summary data frame to hold the results
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=age_names)
purchase_data.head()

#Get rid of duplicate SN
purchase_data_dup=purchase_data.drop_duplicates(subset=["SN"])

#Groupby Age 

age_group=purchase_data_dup.groupby("Age Group")
#print(age_group["Age"].count())

#Percent by age
percent_by_age =(age_group["Age"].count()/ unique_players*100)
#percent_by_age

#Display Age Demographics Table
age_demographics = pd.DataFrame ({"Total Count": age_group["Age"].count(), "Percent of Players":percent_by_age})
age_demographics["Percent of Players"]=age_demographics["Percent of Players"].map("{0:,.2f}%".format)
age_demographics
```

```python
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 46]
names = ["0-9","10-14","15-19","20-24","25-29","30-34","35-39", "40-46"]
#Create a summary data frame to hold the results
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], age_bins, labels=names)
purchase_data.head()

#Groupby Age 

age_group2=purchase_data.groupby("Age Group")

#purchase count
purchase_count_age = age_group2["Age"].count()

#avg purchase price
avg_purchase_price_age = round(age_group2["Price"].mean(),2)


#total purchase price
total_purchase_price_age = age_group2["Price"].sum()

#avg purchase total per person
total_avg_age_pp = round(total_purchase_price_age/age_group["Age"].count(),2)


age_purchase_info = pd.DataFrame ({"Purchase Count": purchase_count_age, "Average Purchase Price":avg_purchase_price_age, "Total Purchase Price": total_purchase_price_age, "Total Average Purchase Price per Person": total_avg_age_pp})
age_purchase_info["Average Purchase Price"]=age_purchase_info["Average Purchase Price"].map("${:.2f}".format)
age_purchase_info["Total Purchase Price"]=age_purchase_info["Total Purchase Price"].map("${:.2f}".format)
age_purchase_info["Total Average Purchase Price per Person"]=age_purchase_info["Total Average Purchase Price per Person"].map("${:.2f}".format)


age_purchase_info
```

```python
#identify top spenders
top_spenders = purchase_data.groupby("SN")

#purchase count of top spenders
purchase_top = top_spenders["Item ID"].count()

#avg purchase price top spenders
avg_purchase_top = top_spenders ["Price"].mean()

#Total purchase value
total_purchase_top = top_spenders ["Price"].sum()

top_spender_info = pd.DataFrame ({"Purchase Count": purchase_top, "Average Purchase Price":avg_purchase_top, "Total Purchase Price":total_purchase_top})
top_spender_info = top_spender_info.sort_values("Total Purchase Price", ascending=False)
top_spender_info["Average Purchase Price"] = top_spender_info["Average Purchase Price"].map("${:.2f}".format)
top_spender_info["Total Purchase Price"] = top_spender_info["Total Purchase Price"].map("${:.2f}".format)
top_spender_info.head(5)
```

```python
#Retrieve the Item ID, Item Name, and Item Price columns
items_df = purchase_data[["Item ID", "Item Name", "Price"]]

#Group by Item ID and Item Name. 
items_df= items_df.groupby(["Item ID", "Item Name"])

#obtain purchase count
purchase_count_items = items_df["Price"].count()


#total purchase value
total_purchase_value = items_df["Price"].sum()

#item price
item_price = total_purchase_value/purchase_count_items

#Create a summary data frame to hold the results

popular_items = pd.DataFrame ({"Purchase Count": purchase_count_items, "Item Price": item_price, "Total Purchase Value":total_purchase_value})

#Sort the purchase count column in descending order
popular_items = popular_items.sort_values("Purchase Count", ascending=False)
popular_items.head(5)
```

```python
#Sort the above table by total purchase value in descending order
popular_items = popular_items.sort_values("Total Purchase Value", ascending=False)
popular_items["Item Price"] = popular_items["Item Price"].map("${:.2f}".format)
popular_items["Total Purchase Value"] = popular_items["Total Purchase Value"].map("${:.2f}".format)
popular_items.head(5)
```
